```python
from PIL import Image
import numpy as np
import pandas as pd
import os
import tensorflow as tf
from tensorflow.keras.layers import Dense,Conv2D,Rescaling,MaxPooling2D,Flatten,Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing import image_dataset_from_directory
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import pathlib 

data_dir =  pathlib.Path('dataset')

image_count = len(list(data_dir.glob('*/*.jpg')))
```

```python
beagle = list(data_dir.glob('Beagle/*'))

beagle_count = len(beagle)

Image.open(str(beagle[2]))
```

```python
# training ImageDataGenerator
train_datagen = ImageDataGenerator(
    rescale=1./255,           # Normalize pixel values
    rotation_range=40,        # Random rotation
    width_shift_range=0.2,    # Horizontal shift
    height_shift_range=0.2,   # Vertical shift
    shear_range=0.2,          # Shear transformation
    zoom_range=0.2,           # Random zoom
    horizontal_flip=True,     # Random horizontal flip
    fill_mode='nearest',      # Fill missing pixels
    validation_split=0.2      # Set aside 20% of data for validation
)
```

```python
# validation ImageDataGenerator
validation_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)
```

```python

batch_size =32
img_height = 180
img_width = 180
```

```python
# creating a dataset 

# generates a dataset from image files in a directory
train_ds = train_datagen.flow_from_directory(data_dir,
                                        subset="training",
                                        seed=123,
                                        class_mode='categorical',
                                        target_size=(img_height,img_width),
                                        batch_size=batch_size,
                                        shuffle=True                              
)

# creating validation dataset
val_ds = validation_datagen.flow_from_directory(data_dir,
                                        subset="validation",
                                        class_mode='categorical',
                                        seed=123,
                                        target_size=(img_height,img_width),
                                        batch_size=batch_size,
                                        shuffle=False
)
```

```python
num_classes =10

model = Sequential()
model.add(Conv2D(32,(3,3),activation='relu'))
model.add(Dropout(0.2))
model.add(MaxPooling2D())
model.add(Conv2D(32,(3,3),activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(32,(3,3),activation='relu'))
model.add(Dropout(0.2))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(num_classes,activation='softmax'))

```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model1=model.fit(train_ds,epochs=20,steps_per_epoch=len(train_ds),batch_size=32,validation_data=val_ds,validation_steps=len(val_ds))
```
